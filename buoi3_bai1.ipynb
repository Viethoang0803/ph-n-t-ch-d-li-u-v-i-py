```python
import random
```

```python
# bai 3
D = {
    'a': 10,
    'b': 25,
    'c': 18,
    'd': 32,
    'e': 8,
}

# Sắp xếp các mục theo giá trị giảm dần và lấy 3 giá trị đầu tiên
sorted_values = sorted(D.values(), reverse=True)[:3]

# In ra 3 giá trị lớn nhất
print("3 giá trị lớn nhất là:", sorted_values)
```

```python
#bai 6
rainfall_data = {}
for month in range(1, 13):
    rainfall_values = [random.uniform(100, 4000) for _ in range(20)]
    rainfall_data[month] = rainfall_values
```

```python
for month, values in rainfall_data.items():
    print(f"Tháng {month}: {values}")
```
