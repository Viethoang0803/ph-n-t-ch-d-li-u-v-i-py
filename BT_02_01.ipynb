### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv',sep = '\t')
chipo.head(5)
```

### Step 4. How many products cost more than $10.00?

```python
chipo[chipo['item_price'].apply(lambda x : float(x.replace('$','')))>10]['item_price'].shape[0]
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_name','item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo[chipo['item_price'].apply(lambda x: float(x.replace('$','')))==chipo['item_price'].apply(lambda x: float(x.replace('$',''))).max()]['quantity'].sum()

```

```python
chipo[chipo['item_price'].apply(lambda x: float(x.replace('$','')))==chipo['item_price'].apply(lambda x: float(x.replace('$',''))).max()]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name']=='Veggie Salad Bowl'][chipo['quantity']==1].shape[0]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo[chipo['item_name']=='Canned Soda'][chipo['quantity']!=1].shape[0]
```
